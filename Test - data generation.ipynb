```python
import numpy as np
import pandas as pd
```

```python
FOLDER = 'data/'
```

```python
n_train = 1000
n_test = 100

for scale in [0.1,1,10,100]:
    train_name = FOLDER+"exp1_data_train_"+str(scale)
    test_name = FOLDER+"exp1_data_test_"+str(scale)

    train_name_s = FOLDER+"exp1_data_train_s_"+str(scale)
    test_name_s = FOLDER+"exp1_data_test_s_"+str(scale)

    var_count = 4
    
    variables = np.array([0.5,0.2,0.8,0.1])

    f = lambda var: var@np.array([[3,1,4,3]]).T

    variable_names = ['one','two','three','four']

    to_add = np.random.rand(n_train,var_count)*100
    noise = np.tile(variables,(n_train,1))*scale

    values = np.array([
        [np.random.normal(0,noise[row,col]) for col in range(var_count)]
    for row in range(n_train)]) + to_add

    values = np.hstack([values,f(to_add)])

    train_data = pd.DataFrame(values,columns=variable_names+['y'])
    train_data.to_csv(train_name+'.csv',index=False)
    train_data_s = pd.DataFrame(noise,columns=variable_names)
    train_data_s.to_csv(train_name_s+'.csv',index=False)


    to_add = np.random.rand(n_test,var_count)*100
    noise = np.tile(variables,(n_test,1))*scale

    values_test = np.array([
        [np.random.normal(0,noise[row,col]) for col in range(var_count)]
    for row in range(n_test)]) + to_add

    values_test = np.hstack([values_test,f(to_add)])

    test_data = pd.DataFrame(values_test,columns=variable_names+['y'])
    test_data.to_csv(test_name+'.csv',index=False)
    test_data_s = pd.DataFrame(noise,columns=variable_names)
    test_data_s.to_csv(test_name_s+'.csv',index=False)
test_data.head()
```

```python
n_train = 1000
n_test = 100

for scale in [0.1,1,10,100]:
    train_name = FOLDER+"exp3_data_train_"+str(scale)
    test_name = FOLDER+"exp3_data_test_"+str(scale)
    
    train_name_s = FOLDER+"exp3_data_train_s_"+str(scale)
    test_name_s = FOLDER+"exp3_data_test_s_"+str(scale)
    
    var_count = 4

    variables = np.array([0.5,0.2,0.8,0.1])

    f = lambda var: var@np.array([[3,1,4,3]]).T

    variable_names = ['one','two','three','four']

    to_add = np.random.rand(n_train,len(variables)-1)
    to_add = np.hstack([to_add,to_add[:,-1,None]])*100
    noise = np.tile(variables,(n_train,1))*scale
    values = np.array([
        [np.random.normal(0,noise[row,col]) for col in range(var_count)]
    for row in range(n_train)]) + to_add

    values = np.hstack([values,f(to_add)])

    train_data = pd.DataFrame(values,columns=variable_names+['y'])
    train_data.to_csv(train_name+'.csv',index=False)
    train_data_s = pd.DataFrame(noise,columns=variable_names)
    train_data_s.to_csv(train_name_s+'.csv',index=False)

    to_add = np.random.rand(n_test,len(variables)-1)
    to_add = np.hstack([to_add,to_add[:,-1,None]])*100
    noise = np.tile(variables,(n_test,1))*scale
    values_test = np.array([
        [np.random.normal(0,noise[row,col]) for col in range(var_count)]
    for row in range(n_test)]) + to_add

    values_test = np.hstack([values_test,f(to_add)])

    test_data = pd.DataFrame(values_test,columns=variable_names+['y'])
    test_data.to_csv(test_name+'.csv',index=False)
    test_data_s = pd.DataFrame(noise,columns=variable_names)
    test_data_s.to_csv(test_name_s+'.csv',index=False)
test_data.head()
```
